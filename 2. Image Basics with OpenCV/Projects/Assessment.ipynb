```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
plt.imshow(pic)
```

```python
fix_pic = cv2.cvtColor(pic,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_pic)
```

```python
plt.imshow(flip_pic)
```

```python
fix_pic = cv2.cvtColor(pic,cv2.COLOR_BGR2RGB)
plt.imshow(fix_pic)
```

```python
cv2.rectangle(fix_pic,(200,370),(600,700),(255,0,0),thickness=10)
plt.imshow(fix_pic)
```

```python
# you can use polyLines here too
fix_pic = cv2.cvtColor(pic,cv2.COLOR_BGR2RGB)
pts = np.array([[430,400],[200,700],[600,700]])
cv2.drawContours(fix_pic,[pts],0,(0,0,255),thickness=20)
plt.imshow(fix_pic)
```

```python
# you can use polyfill here too
fix_pic = cv2.cvtColor(pic,cv2.COLOR_BGR2RGB)
pts = np.array([[430,400],[200,700],[600,700]])
cv2.drawContours(fix_pic,[pts],0,(0,0,255),thickness=-1)
plt.imshow(fix_pic)
```

```python
fix_pic = cv2.cvtColor(pic,cv2.COLOR_BGR2RGB)
pts = np.array([[430,400],[200,700],[600,700]])
pts = pts.reshape((-1,1,2))
cv2.polylines(fix_pic,[pts],isClosed=True,color=(255,0,0),thickness=10)
plt.imshow(fix_pic)
```

```python
fix_pic = cv2.cvtColor(pic,cv2.COLOR_BGR2RGB)
pts = np.array([[430,400],[200,700],[600,700]])
pts = pts.reshape((-1,1,2))
cv2.fillPoly(fix_pic,[pts],color=(255,0,0))
plt.imshow(fix_pic)
```
